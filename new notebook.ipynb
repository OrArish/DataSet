```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/OrArish/DataSet/main/Sleep_Efficiency.csv"
sleep_df = pd.read_csv(url)
sleep_df.info()
```

```python
fig, ax = plt.subplots (figsize=(15, 5),  ncols=10)
plt.subplots_adjust(wspace = 0.3)
sns.boxplot(data = sleep_df[['Age']], ax = ax[0])
sns.boxplot(data = sleep_df[['REM sleep percentage']], ax = ax[1])
sns.boxplot(data = sleep_df[['Sleep duration']], ax = ax[2])
sns.boxplot(data = sleep_df[['Sleep efficiency']], ax = ax[3])
sns.boxplot(data = sleep_df[['Deep sleep percentage']], ax = ax[4])
sns.boxplot(data = sleep_df[['Light sleep percentage']], ax = ax[5])
sns.boxplot(data = sleep_df[['Awakenings']], ax = ax[6])
sns.boxplot(data = sleep_df[['Caffeine consumption']], ax = ax[7])
sns.boxplot(data = sleep_df[['Alcohol consumption']], ax = ax[8])
sns.boxplot(data = sleep_df[['Exercise frequency']], ax = ax[9])
plt.show
```

```python
sleep_df["Smoking status"] = sleep_df["Smoking status"].replace({'Yes': 1, 'No' : 0})
```

```python
data = {sleep_df['Age'],sleep_df[['REM sleep percentage']],sleep_df[['Sleep duration']],sleep_df[['Sleep efficiency']],}
corr_matrix = sleep_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Convert 'Bedtime' and 'Wakeup time' to datetime
sleep_df['Bedtime'] = pd.to_datetime(sleep_df['Bedtime'])
sleep_df['Wakeup time'] = pd.to_datetime(sleep_df['Wakeup time'])

# Convert 'Smoking status' from 'Yes'/'No' to 1/0
sleep_df['Smoking status'] = sleep_df['Smoking status'].replace({'Yes': 1, 'No': 0})

# Display data types to ensure proper conversion
sleep_df.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics
summary_stats = sleep_df.describe()

# Plot distributions of key variables
plt.figure(figsize=(15, 10))
sns.histplot(sleep_df['Age'], kde=True)
plt.title('Distribution of Age')
plt.show()

plt.figure(figsize=(15, 10))
sns.histplot(sleep_df['Sleep duration'], kde=True)
plt.title('Distribution of Sleep Duration')
plt.show()

plt.figure(figsize=(15, 10))
sns.histplot(sleep_df['Sleep efficiency'], kde=True)
plt.title('Distribution of Sleep Efficiency')
plt.show()
```

```python
# Check unique values in 'Exercise frequency' column
unique_exercise_values = sleep_df['Exercise frequency'].unique()
unique_exercise_values
```

```python
# Convert 'Exercise frequency' to numeric, coercing errors
sleep_df['Exercise frequency'] = pd.to_numeric(sleep_df['Exercise frequency'], errors='coerce')

# Drop rows with NaN values in 'Exercise frequency' (if any)
sleep_df.dropna(subset=['Exercise frequency'], inplace=True)

# Plot the relationship between 'Exercise frequency' and 'Deep sleep percentage'
plt.figure(figsize=(10, 6))
sns.boxplot(data=sleep_df, x='Exercise frequency', y='Deep sleep percentage')
plt.title('Exercise Frequency vs. Deep Sleep Percentage')
plt.show()
```
